```python
# Source - https://stackoverflow.com/a
# Posted by jacquesdirac
# Retrieved 2025-11-18, License - CC BY-SA 4.0

from tensorboard.backend.event_processing.event_accumulator import EventAccumulator
    
event_acc = EventAccumulator(log_dir_train, size_guidance={"scalars": 50})
event_acc.Reload()

for scalar in event_acc.Tags()["scalars"]:
    print(scalar)
    # w_times, step_nums, vals = zip(*event_acc.Scalars(scalar))
```
